```python
# data = [("1", "Spark NLP is an awesome library. Natural language processing made easy with Spark."),
#                # 'B-PROD, I-PROD, O, O, B-FEAT, B-PROD, O, B-PROD, I-PROD, I-PROD, O, O, O, B-PROD, O'),
#         ("2", "Natural language processing made easy with Spark. Spark NLP is an awesome library."),
#                 # 'B-PROD, I-PROD, I-PROD, O, O, O, B-PROD, O, B-PROD, I-PROD, O, O, B-FEAT, B-PROD, O')
#        ]

data = [["Spark NLP is an awesome library. Natural language processing made easy with Spark."]]
```

```python
df = spark.createDataFrame(data).toDF('text')
```

```python
df.show(truncate=False)
```

```python
document = DocumentAssembler() \
            .setInputCol('text') \
            .setOutputCol('document') \
            .setCleanupMode('shrink') # this removes extra spaces
```

```python
sentence = SentenceDetector() \
            .setInputCols('document') \
            .setOutputCol('sentence')
```

```python
sentence.setExplodeSentences(True) 
# this will put each of the sentences on a new df row 
# which will increase the parallelisation of our spark dataframe -> increased speed
```

```python
tokenizer = Tokenizer() \
            .setInputCols('sentence') \
            .setOutputCol('token')
# tokenizer.set() # to be read in documentation
tokenizer.setExceptions(['e-mail', 'email'])
```

```python
# spell checker NorvigSweetingModel or SymmetricDelete

checker = NorvigSweetingModel.pretrained() \
            .setInputCols(['token']) \
            .setOutputCol('checked')
```

```python
embeddings = WordEmbeddingsModel.pretrained() \
                        .setInputCols(["sentence", "checked"]) \
                        .setOutputCol("embeddings") \
                        .setCaseSensitive(False)
```

```python
ner = NerDLModel.pretrained() \
            .setInputCols(['sentence', 'checked', "embeddings"]) \
            .setOutputCol('ner')
```

```python
# create the chunks
converter = NerConverter().setInputCols(['sentence', 'checked', 'ner']).setOutputCol('chunk')
```

```python
from pyspark.ml import Pipeline
```

```python
stages = [
    document,
    sentence,
    tokenizer,
    checker,
    embeddings,
    ner,
    # converter,
]

pipeline = Pipeline().setStages(stages)
```

```python
model = pipeline.fit(df)
```

```python
# sorted(model.__dir__())
model.transform(df)
```
